### Disney Characters

```python
disney = requests.get("https://api.disneyapi.dev/character")
print("Disney Characters:", disney.status_code)
```

```python
result = disney.json()
#result
```

#### Brute forcing it to work

```python
flattened_data = json_normalize(result)

flattened_data
```

```python
flattened_data2 = json_normalize(flattened_data['data']).T
flattened_data2
```

```python
flattened_data3 = json_normalize(flattened_data2[0])
flattened_data3
```

#### Directly getting the data dictionary in a df

```python
results4 = disney.json()
results4['data']

pd.set_option("display.max_rows", 0)

data4 = pd.DataFrame(results4['data'])
data4
```
